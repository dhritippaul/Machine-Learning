```python
labels = 'Python' , 'C++' , 'Ruby' , 'Java'
sizes = [215,130,245,210]
colours = ['gold' , 'yellow','red','blue']
explode = (0.1,0,0,0)


plt.pie(
   sizes,
    explode=explode,
    labels=labels,
    colors=colours,
    autopct='%1.1f%%',
    shadow = True
)
plt.axis('equal')
```
