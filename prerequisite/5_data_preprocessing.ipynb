```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Life_expectency.csv")
```

```python
n_col = df.select_dtypes(include=np.number).columns
print("Numerical columns : \n",n_col)
```

```python
c_col = df.select_dtypes(exclude= np.number).columns
print("Categorical columns: \n",c_col)
```

```python
df = df.rename(columns = lambda x : x.strip())
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Status']=label_encoder.fit_transform(df['Status'])
df.head()
```

```python
print(df.isna().sum())
print(df.shape)
```

```python
#drop.treat.algo
for i in df.columns.drop('Life expectancy'):
    df[i].fillna(df[i].mean, inplace = True)
```

```python
plt.figure(figsize=(8,8),dpi=80)
sns.boxplot(df['Life expectancy'])
plt.title('Life expectancy box plot')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title('Life expectancy')
sns.distplot(df['Life expectancy'])
```

```python
num_col = df.select_dtypes(include=np.number).columns
print("Numerical columns: \n",num_col)

cat_col = df.select_dtypes(exclude=np.number).columns
print("Categorical columns: \n",cat_col)
```

```python
# Let's check the multicollinearity of features by checking the correlation matric

plt.figure(figsize=(15,15))
p=sns.heatmap(df[num_col].corr(), annot=True,cmap='RdYlGn',center=0)
```
