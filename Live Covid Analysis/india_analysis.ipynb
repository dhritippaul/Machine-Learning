```python
import pandas as pd
```

```python
dpp = pd.read_json('datanew.json')
dpp
```

```python
dpp = dpp.iloc[:-1 , 1:]
```

```python
dpp.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(dpp['state_name']))
total_states
```

```python
dpp['active']
```

```python
max(dpp['positive'])
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,dpp['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, dpp['state_name'])  
plt.xlim(1,max(dpp['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,dpp['new_active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, dpp['state_name'])  
plt.xlim(1,max(dpp['new_active'])+10)
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,dpp['death'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, dpp['state_name'])  
plt.xlim(1,max(dpp['death'])+10) 
plt.xlabel('Death Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```
