```python
frame = cv2.imread("download.jfif")
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
face = rgb[y:y+h,x:x+h]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),2)
```

```python
plt.imshow(out)
```
