### ***This exercise illustrates how the human vision system can identify the outline of the objects in a scene even from low quality images.***

**Step 1**. Create a DoG filter to emulate lateral inhibition. Apply it on the images and see the results. Experiment with different values of the parameters.

**Step 2**. Create oriented filters (0⁰, 45⁰, 90⁰, 135⁰) to detect oriented edges on the DoG filtered image. Examine the edge images

**Step 3.** Superimpose the edge images (applying  WTA policy and normalization) to reproduce the object outlines. Comment on your findings.


```python
import cv2 #importing openCv library
import numpy as np #importing numpy as np
from matplotlib import pyplot as plt #importing pyplot for plotting from matplotlib
from google.colab.patches import cv2_imshow #cv2_imshow specifically for collab
```

```python
peak = cv2.imread('/Peak.jpg') #reading/taking input image
cv2_imshow(peak) #showing image
```

```python
drama = cv2.imread('/Drama.jpg')#reading/taking input
cv2_imshow(drama)#displaying output
```

```python
resize_drama = cv2.resize(drama, (600,400)) #voluntarily resizing for better view in collab
```

```python
resize_peak = cv2.resize(peak, (600,400))
```

```python
drama_new = cv2.GaussianBlur(resize_drama,(1,1),cv2.BORDER_DEFAULT) #Applying GaussianBlur and changing standard deviation parameter in drama iamge
drama_new2 = cv2.GaussianBlur(resize_drama,(3,3),cv2.BORDER_DEFAULT)
drama_new3 = cv2.GaussianBlur(resize_drama,(7,7),cv2.BORDER_DEFAULT)
```

```python
peak_new = cv2.GaussianBlur(resize_peak,(1,1),cv2.BORDER_DEFAULT)#Applying GaussianBlur and changing standard deviation parameter in peak image
peak_new2 = cv2.GaussianBlur(resize_peak,(3,3),cv2.BORDER_DEFAULT)
peak_new3 = cv2.GaussianBlur(resize_peak,(7,7),cv2.BORDER_DEFAULT)
```

```python
print("Resized original image :-") #representing the images 
cv2_imshow(resize_drama)
print("  ")

print("Image after 1st processing :-")
cv2_imshow(drama_new)
print("  ")

print("Image after 2nd processing  :-")
cv2_imshow(drama_new2)
print("  ")

print("Image after 3rd processing  :-")
cv2_imshow(drama_new3)
```

```python
print("Resized original image :-")
cv2_imshow(resize_peak)
print("  ")

print("Image after 1st processing :-")
cv2_imshow(peak_new)
print("  ")

print("Image after 2nd processing  :-")
cv2_imshow(peak_new2)
print("  ")

print("Image after 3rd processing  :-")
cv2_imshow(peak_new3)
```

```python
dog = drama_new - resize_drama #calculating Difference of Gaussian on the above images 
dog2 = drama_new2 - resize_drama
dog3 = drama_new3 - resize_drama
```

```python
dogp = peak_new - resize_peak
dogp2 = peak_new2 - resize_peak
dogp3 = peak_new3 - resize_peak
```

```python
print("DoG @1st processed image :-")#representing DoG images
cv2_imshow(dog)
print("  ")
print("DoG @2nd processed image :-")
cv2_imshow(dog2)
print("  ")
print("DoG @3rd processed image :-")
cv2_imshow(dog3)
```

```python
print("DoG @1st processed image :-")
cv2_imshow(dogp)
print("  ")
print("DoG @2nd processed image :-")
cv2_imshow(dogp2)
print("  ")
print("DoG @3rd processed image :-")
cv2_imshow(dogp3)
```

```python
print("Edge detection on dog2 :-") #finding the edges and changing the parameters for better result on dog2 image
print("Parameter 1")
edge = cv2.Canny(dog2,1000,1000)
cv2_imshow(edge)
print(" ")
print("Parameter 2")
edge = cv2.Canny(dog2,10,10)
cv2_imshow(edge)
```

```python
print("Edge detection on dog2 :-")#finding the edges and changing the parameters for better result on dogp2 image
print("Parameter 1")
edgep = cv2.Canny(dogp2,1000,1000)
cv2_imshow(edgep)
print(" ")
print("Parameter 2")
edgep = cv2.Canny(dogp2,10,10)
cv2_imshow(edgep)

```

```python
print("Edge detection on dog3 :-") #oriented edge detection on dog3 image
print("Parameter 1")
edge1 = cv2.Canny(dog3,1000,1000)
cv2_imshow(edge1)
print(" ")
print("Parameter 2")
edge = cv2.Canny(dog3,10,10)
cv2_imshow(edge)
```

```python
print("Edge detection on dog3 :-")
print("Parameter 1")
edgep1 = cv2.Canny(dogp3,1000,1000)
cv2_imshow(edgep1)
print(" ")
print("Parameter 2")
edgep = cv2.Canny(dogp3,10,10)
cv2_imshow(edgep)
```

```python
print("0 degree rotation : ") #rotation of images at 0 degree
h,w = edge1.shape
center = (h//2,w//2)
rotate = cv2.getRotationMatrix2D(center,0,1)
final = cv2.warpAffine(edge1,rotate,(w,h))
cv2_imshow(final)


print("0 degree rotation : ")
hp,wp = edgep1.shape
centerp = (h//2,w//2)
rotatep = cv2.getRotationMatrix2D(centerp,0,1)
finalp = cv2.warpAffine(edgep1,rotatep,(wp,hp))
cv2_imshow(finalp)
```

```python
print("45 degree rotation : ")#rotation of images at 45 degree
h2,w2 = edge1.shape
center2 = (h2//2,w2//2)
rotate2 = cv2.getRotationMatrix2D(center2,45,1)
final2 = cv2.warpAffine(edge1,rotate2,(w2,h2))
cv2_imshow(final2)

print("45 degree rotation : ")
hp2,wp2 = edgep1.shape
centerp2 = (hp2//2,wp2//2)
rotatep2 = cv2.getRotationMatrix2D(centerp2,45,1)
finalp2 = cv2.warpAffine(edgep1,rotatep2,(wp2,hp2))
cv2_imshow(finalp2)
```

```python
print("90 degree rotation : ")#rotation of images at 90 degree
h3,w3 = edge1.shape
center3 = (h3//2,w3//2)
rotate3 = cv2.getRotationMatrix2D(center3,90,0.8)
final3 = cv2.warpAffine(edge1,rotate3,(w3,h3))
cv2_imshow(final3)

print("90 degree rotation : ")
hp3,wp3 = edgep1.shape
centerp3 = (hp3//2,wp3//2)
rotatep3 = cv2.getRotationMatrix2D(centerp3,90,0.8)
finalp3 = cv2.warpAffine(edgep1,rotatep3,(wp3,hp3))
cv2_imshow(finalp3)
```

```python
print("135 degree rotation : ")#rotation of images at 135 degree
h4,w4 = edge1.shape
center4 = (h4//2,w4//2)
rotate4 = cv2.getRotationMatrix2D(center4,135,0.8)
final4 = cv2.warpAffine(edge1,rotate4,(w4,h4))
cv2_imshow(final4)

print("135 degree rotation : ")
hp4,wp4 = edgep1.shape
centerp4 = (hp4//2,wp4//2)
rotatep4 = cv2.getRotationMatrix2D(centerp4,135,0.8)
finalp4 = cv2.warpAffine(edgep1,rotatep4,(wp4,hp4))
cv2_imshow(finalp4)
```

```python
print("Superimposed edge images to reproduce the object outline :- ") #superimposing with dog2 images of drama and peak
add2 = cv2.addWeighted(dog2, 0.5,drama_new2, 0.5,0)
cv2_imshow(add2)

print("Superimposed edge images to reproduce the object outline :- ")
addp2 = cv2.addWeighted(dogp2, 0.5,peak_new2, 0.5,0)
cv2_imshow(addp2)
```

```python
print("Superimposed edge images to reproduce the object outline :- ") #superimposing with dog3 images of drama and peak
add1 = cv2.addWeighted(dog3, 0.5,drama_new3, 0.5,0)
cv2_imshow(add1)

print("Superimposed edge images to reproduce the object outline :- ")
addp1 = cv2.addWeighted(dogp3, 0.5,peak_new3, 0.5,0)
cv2_imshow(addp1)
```
